```python
# Preprocess test dataset
train_processed = train_data.copy()

# Drop identifier columns
train_processed.drop(columns=['Item_Identifier', 'Outlet_Identifier'], inplace=True)

# One-hot encode categorical columns
train_categorical_cols = train_processed.select_dtypes(include=['object']).columns
train_processed = pd.get_dummies(train_processed, columns=train_categorical_cols)

train_processed.head()
```

```python
# Preprocess test dataset
test_processed = test_data.copy()

# Drop identifier columns
test_processed.drop(columns=['Item_Identifier', 'Outlet_Identifier'], inplace=True)

# One-hot encode categorical columns
test_categorical_cols = test_processed.select_dtypes(include=['object']).columns
test_processed = pd.get_dummies(test_processed, columns=test_categorical_cols)

test_processed.head()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

correlation_matrix = train_processed.corr()

# Outlet Size sütunları için mutlak korelasyonları sıralama
outlet_high_corr = correlation_matrix['Outlet_Size_High'].abs().sort_values(ascending=False).head(10)
outlet_medium_corr = correlation_matrix['Outlet_Size_Medium'].abs().sort_values(ascending=False).head(10)
outlet_small_corr = correlation_matrix['Outlet_Size_Small'].abs().sort_values(ascending=False).head(10)

# Her biri için grafikleri çizme
fig, axes = plt.subplots(3, 1, figsize=(8, 12))

outlet_high_corr.plot(kind='barh', ax=axes[0], color='blue', title='Top 5 Correlations for Outlet_Size_High')
axes[0].set_xlabel('Correlation')

outlet_medium_corr.plot(kind='barh', ax=axes[1], color='green', title='Top 5 Correlations for Outlet_Size_Medium')
axes[1].set_xlabel('Correlation')

outlet_small_corr.plot(kind='barh', ax=axes[2], color='red', title='Top 5 Correlations for Outlet_Size_Small')
axes[2].set_xlabel('Correlation')

plt.tight_layout()
plt.show()
```

```python
train_copy = train_data.copy()
train_categorical_columns_copy = train_copy.select_dtypes(include=['object']).columns
```

```python
# Option 2: Label encoding
from sklearn.preprocessing import LabelEncoder

categorical_columns = train_copy.select_dtypes(include=['object']).columns
numeric_column = 'Item_Weight'

le = LabelEncoder()
for col in train_categorical_columns_copy:
    train_copy[col + '_encoded'] = le.fit_transform(train_copy[col])

# Korelasyon matrisini hesapla
correlation_matrix = train_copy[[numeric_column] + [col + '_encoded' for col in train_categorical_columns_copy]].corr()

# Sayısal sütunun diğer sütunlarla olan korelasyonunu al
correlation_values = correlation_matrix[numeric_column]

# Korelasyon değerlerini mutlak değere göre sıralayıp, en yüksek 5 değeri seç
top_5_correlation = correlation_values.abs().sort_values(ascending=False).head(5)

# En iyi 5 korelasyon değerini yazdır
print(top_5_correlation)
```

```python
item_type_list_test = test_data['Item_Identifier'].unique()
item_type_list_train = train_data['Item_Identifier'].unique()
```

```python
overall_mode_test = test_data['Item_Weight'].mode().iloc[0]  # Genel mod değeri

for type in item_type_list_test:
    mode_value = test_data[test_data['Item_Identifier'] == type]['Item_Weight'].mode()
    if not mode_value.empty:  # Grup mod değeri mevcutsa
        test_data.loc[(test_data['Item_Weight'].isna()) & (test_data['Item_Identifier'] == type), 'Item_Weight'] = mode_value.iloc[0]
    else:  # Grup için mod yoksa genel mod değeri kullanılır
        test_data.loc[(test_data['Item_Weight'].isna()) & (test_data['Item_Identifier'] == type), 'Item_Weight'] = overall_mode_test
```

```python
overall_mode_train = train_data['Item_Weight'].mode().iloc[0]  # Genel mod değeri

for type in item_type_list_train:
    mode_value = train_data[train_data['Item_Identifier'] == type]['Item_Weight'].mode()
    if not mode_value.empty:  # Grup mod değeri mevcutsa
        train_data.loc[(train_data['Item_Weight'].isna()) & (train_data['Item_Identifier'] == type), 'Item_Weight'] = mode_value.iloc[0]
    else:  # Grup için mod yoksa genel mod değeri kullanılır
        train_data.loc[(train_data['Item_Weight'].isna()) & (train_data['Item_Identifier'] == type), 'Item_Weight'] = overall_mode_train
```
