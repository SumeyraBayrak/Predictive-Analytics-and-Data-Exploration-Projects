```python
def convert_indian_units(series):
    """
    Convert numbers with Indian units (Crore/Lakh) to their decimal equivalents.
    
    Args:
        series: Pandas Series containing strings like "10 Cr" or "50 Lakh"
        
    Returns:
        Pandas Series with converted decimal values
        Returns pd.NA for invalid or unrecognized formats
        
    Examples:
        "10 Cr" -> 1000 (10 * 100)
        "50 Lakh" -> 50
        "Invalid" -> pd.NA
    """

    def _convert_single_value(value):
        # Handle null/empty values
        if pd.isna(value):
            return np.nan
            
        try:
            # Clean and split the input string
            value_str = str(value).strip()
            parts = value_str.split()
            
            # Check if we have both number and unit
            if len(parts) != 2:
                return np.nan
                
            # Extract number and unit
            number = float(parts[0])
            unit = parts[1].lower()
            
            # Convert based on unit
            if unit in ['cr', 'crore']:
                return number * 100  # 1 Crore = 100 Lakhs
            elif unit in ['lakh', 'lac']:
                return number  # Keep Lakh as is
            else:
                return np.nan  # Unrecognized unit
                
        except (ValueError, TypeError, AttributeError):
            # Return NA for any conversion errors
            return np.nan
    
    # Apply conversion to each value in the series
    return series.apply(_convert_single_value)
```

```python
def convert_new_price(new_price):
    if pd.isna(new_price): 
        return new_price
    
    price_str = str(new_price)
    
    if 'Lakh' in price_str:
        value = float(price_str.split()[0]) * 100  # 1 Lakh = 100 K INR
    elif 'Cr' in price_str:
        value = float(price_str.split()[0]) * 10_000  # 1 Crore = 10,000 K INR
    else:
        return new_price

    return value
```
