```python
def convert_new_price(new_price):
    if pd.isna(new_price): 
        return new_price
    
    price_str = str(new_price)
    
    if 'Lakh' in price_str:
        value = float(price_str.split()[0]) * 100  # 1 Lakh = 100 K INR
    elif 'Cr' in price_str:
        value = float(price_str.split()[0]) * 10_000  # 1 Crore = 10,000 K INR
    else:
        return new_price

    return value
```
