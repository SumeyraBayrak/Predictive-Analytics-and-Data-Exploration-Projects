# üö≤ Predicting London‚Äôs Bike Shares  

**Goal**: Forecast hourly bike rentals using weather, holidays, and seasons!  

---

## üå¶Ô∏è Where‚Äôs the Data From?  

1. **TfL Bike Shares**  
   - *Source*: [TfL Cycling Data](https://cycling.data.tfl.gov.uk)  
   - *What‚Äôs included?* Hourly bike rentals from **2015‚Äì2016** (short trips only).  
   - *Credits*: `Powered by TfL Open Data` + `Contains OS data ¬© Crown copyright`  

2. **Weather Mood** ‚òî  
   - *Source*: Freemeteo.com  
   - *Details*: Real temp (¬∞C), "feels like" temp, humidity, wind speed, and weather codes (sun, rain, snow‚Ä¶ even freezing fog!).  

3. **Holidays & Weekends** üìÖ  
   - *Source*: UK Government‚Äôs holiday calendar.  
   - *Why?* Fewer commuters on holidays/weekends = fewer bikes rented!  

---

## üîç What‚Äôs Inside the Data?  

- **`timestamp`**: Hourly timestamps (e.g., *Jan 1, 2015 @ 8AM*).  
- **`cnt`**: Number of bikes rented that hour.  
- **`t1` vs `t2`**: Real temperature vs. "feels like" (ever dressed wrong for the weather? üòÖ).  
- **`hum`**: Humidity (% ‚Äî bad hair days matter!).  
- **`weather_code`**: Simplified weather types:  
  - 1 = Clear üåû | 7 = Rain üåßÔ∏è | 26 = Snow ‚ùÑÔ∏è | 94 = Freezing Fog (ü•∂ + üå´Ô∏è).  
- **`is_holiday`**: 1 = Holiday üéâ | 0 = Normal day.  
- **`season`**: 0 (üå∏ Spring) ‚Üí 3 (‚ùÑÔ∏è Winter).

```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns

# Set a modern, clean style for our plots
sns.set(style="whitegrid", context="talk")

# Assume df has the columns: "weekday", "hour", and "count"
# Compute the average bike rentals for each combination of weekday and hour
avg_demand = data.groupby(["weekday", "hour"])["cnt"].mean().unstack("weekday")

# Optionally, ensure weekdays are ordered correctly.
# Here we assume weekday numbers: 0=Sun, 1=Mon, 2=Tue, 3=Wed, 4=Thu, 5=Fri, 6=Sat.
weekday_order = [0, 1, 2, 3, 4, 5, 6]
avg_demand = avg_demand[weekday_order]

# Rename the columns to weekday names for a clearer legend
weekday_labels = {0: "Sun", 1: "Mon", 2: "Tue", 3: "Wed", 4: "Thu", 5: "Fri", 6: "Sat"}
avg_demand.rename(columns=weekday_labels, inplace=True)

# Create the plot
plt.figure(figsize=(12, 6))

# Plot a line for each weekday
for day in avg_demand.columns:
    plt.plot(
        avg_demand.index,         # Hours of the day (0-23)
        avg_demand[day],          # Average rentals for this day at each hour
        marker="o",               # Add markers to each data point for clarity
        label=day                 # Use the weekday name as label
    )

# Customize the plot with titles, labels, and ticks
plt.title("Average Hourly Bike Demand by Weekday", fontsize=16, fontweight="bold")
plt.xlabel("Hour of the Day", fontsize=14)
plt.ylabel("Average Number of Bike Rentals", fontsize=14)
plt.xticks(range(0, 24, 1))  # Show x-axis ticks every 2 hours

# Add a legend to indicate which line corresponds to which weekday
plt.legend(title="Weekday")

plt.tight_layout()
plt.show()

```

```python
weekend_busy_hours = range(11, 19) 
morning_busy_hours = [7, 8, 9]
evening_busy_hours = [16, 17, 18, 19]

data['busy_category'] = np.select(
    [
        # Weekend busy (11am-6pm)
        (data['is_weekend'] == True) & (data['hour'].isin(weekend_busy_hours)),
        # Weekday morning rush (7am-9am)
        (data['is_weekend'] == False) & (data['hour'].isin(morning_busy_hours)),
        # Weekday evening rush (4pm-7pm)
        (data['is_weekend'] == False) & (data['hour'].isin(evening_busy_hours))
    ],
    [
        'weekend_busy',
        'morning_busy',
        'evening_busy'
    ],
    default='not_busy'
)
```

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from scipy import stats

# -----------------------------
# 1. Compute the Correlation Matrix
# -----------------------------
# Identify numeric columns in the DataFrame 'data'
numeric_columns = data.select_dtypes(include=['number']).columns

# Compute and round the correlation matrix to 3 decimal places
correlation_matrix = data[numeric_columns].corr().round(3)

# -----------------------------
# 2. Reshape the Correlation Matrix to Long Format
# -----------------------------
# Reset the index and melt the correlation matrix for easier manipulation
correlation_df = correlation_matrix.reset_index().melt(
    id_vars=['index'],
    var_name='Kar≈üƒ±la≈ütƒ±rƒ±lan √ñzellik',
    value_name='Korelasyon Deƒüeri'
)
# Rename the index column to '√ñzellikler'
correlation_df.rename(columns={'index': '√ñzellikler'}, inplace=True)

# -----------------------------
# 3. Enhance the Correlation DataFrame
# -----------------------------
# Filter out self-comparisons (where a feature is compared with itself)
correlation_analysis = correlation_df.query('√ñzellikler != `Kar≈üƒ±la≈ütƒ±rƒ±lan √ñzellik`').copy()

# Calculate the absolute correlation value
correlation_analysis['Mutlak_Korelasyon'] = correlation_analysis['Korelasyon Deƒüeri'].abs()

# Create a feature pair identifier. This ensures that the pair A ‚Üî B is the same as B ‚Üî A.
condition = correlation_analysis['√ñzellikler'] < correlation_analysis['Kar≈üƒ±la≈ütƒ±rƒ±lan √ñzellik']
correlation_analysis['√ñzellik_√áifti'] = np.where(
    condition,
    correlation_analysis['√ñzellikler'] + ' ‚Üî ' + correlation_analysis['Kar≈üƒ±la≈ütƒ±rƒ±lan √ñzellik'],
    correlation_analysis['Kar≈üƒ±la≈ütƒ±rƒ±lan √ñzellik'] + ' ‚Üî ' + correlation_analysis['√ñzellikler']
)

# Calculate the explained variance (in percentage) based on the correlation
correlation_analysis['A√ßƒ±klanan_Varyans'] = (correlation_analysis['Korelasyon Deƒüeri'] ** 2 * 100).round(1)

# Categorize the relationship strength based on absolute correlation
bins = [0, 0.2, 0.4, 0.6, 0.8, 1]
labels = ['√áok Zayƒ±f', 'Zayƒ±f', 'Orta', 'G√º√ßl√º', '√áok G√º√ßl√º']
correlation_analysis['ƒ∞li≈üki_G√ºc√º'] = pd.cut(
    correlation_analysis['Mutlak_Korelasyon'],
    bins=bins,
    labels=labels,
    right=False
)

# -----------------------------
# 4. Select the Top 10 Unique Correlation Pairs
# -----------------------------
# Remove duplicate feature pairs, sort by the absolute correlation, and select the top 10
unique_correlations = (
    correlation_analysis
    .drop_duplicates('√ñzellik_√áifti')
    .sort_values('Mutlak_Korelasyon', ascending=False)
    .head(10)
    .reset_index(drop=True)
)

# -----------------------------
# 5. Display the Results with Styling
# -----------------------------
# Subset the desired columns and apply a background gradient on 'Mutlak_Korelasyon'
display(
    unique_correlations[['√ñzellik_√áifti', 'Korelasyon Deƒüeri', 'Mutlak_Korelasyon', 'A√ßƒ±klanan_Varyans', 'ƒ∞li≈üki_G√ºc√º']]
    .style
    .background_gradient(subset=['Mutlak_Korelasyon'], cmap='ocean')
    .format({
        'Korelasyon Deƒüeri': '{:.3f}',
        'Mutlak_Korelasyon': '{:.3f}',
        'A√ßƒ±klanan_Varyans': '{:.1f}%',
    })
)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# G√∂rselle≈ütirme i√ßin fig√ºr ve eksen olu≈ütur
fig, ax = plt.subplots(figsize=(12, 6))

sns.barplot(
    x='Korelasyon Deƒüeri',  
    y='√ñzellik_√áifti', 
    data=unique_correlations,  
    hue='ƒ∞li≈üki_G√ºc√º',  
    palette="ocean",  # Belirlediƒüimiz renkler kullanƒ±lƒ±yor
    dodge=False,  
    edgecolor='white',  
    ax=ax
)

# x=0 referans √ßizgisi
ax.axvline(0, color='black', linestyle='--', alpha=0.7)

# Yatay grid √ßizgileri ekleyelim
ax.grid(axis='x', linestyle='--', alpha=0.3, color='gray')

# Grafik ba≈ülƒ±ƒüƒ± ve etiketler
ax.set_title('En G√º√ßl√º 10 √ñzellik ƒ∞li≈ükisi', pad=15, fontsize=16)
ax.set_xlabel('Korelasyon Deƒüeri', fontsize=14)
ax.set_ylabel('√ñzellik √áiftleri', fontsize=14)

# Legend pozisyonunu ayarlayalƒ±m
ax.legend(title='ƒ∞li≈üki G√ºc√º', bbox_to_anchor=(1.05, 1), loc='upper left')

# Grafik d√ºzenini sƒ±kƒ±≈ütƒ±rarak kenar ta≈ümalarƒ±nƒ± √∂nleyelim
plt.tight_layout()

# Grafiƒüi g√∂ster
plt.show()
```
